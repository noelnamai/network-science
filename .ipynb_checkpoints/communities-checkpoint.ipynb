```python
%matplotlib inline

import operator
import pandas as pd
import networkx as nx
import itertools as it
import matplotlib.pyplot as plt

plt.style.use(["seaborn-whitegrid"])
plt.rcParams["figure.figsize"] = 7.5, 5 
```

```python
M = 10
N = 10**4
graph = nx.barabasi_albert_graph(n=N, m=M)
```

```python
G = graph.copy()
result  = []
removed = []
coefficients = nx.clustering(G)

for key in sorted(coefficients, key=coefficients.get, reverse=True):
    G.remove_node(key)
    removed.append(key)
    try:
        giant = max(nx.connected_component_subgraphs(G), key=len)
        result.append({"fraction": len(removed)/N, "size": len(giant)})
    except:
        pass
    
df = pd.DataFrame(result)
df.plot(x="fraction", y="size", kind="line", legend=False)
plt.title("Removing the node with the \n highest clustering coefficient from the Barabasi-Albert model.")
plt.xlabel(r"Fraction nodes removed $(S/N)$", fontsize=16)
plt.ylabel(r"Size of largest component $(C)$", fontsize=16)
#plt.plot()
plt.savefig("pica.png")  
plt.close()
```

```python
def generate_clique(n=5):
    graph = nx.complete_graph(n)
    for n in graph.nodes():
        if n == 0:
            graph.node[n]["peripheral"] = False
        else:
            graph.node[n]["peripheral"] = True
    return graph
```

```python
graph = generate_clique()
positions = {0: (0,0), 1:(1,1), 2:(-1,-1), 3:(1,-1), 4:(-1,1)}
nx.draw_networkx(graph, pos=positions)
```

```python
def replicate(graph):
    union = graph.copy()
    for i in range(4):
        copy  = graph.copy()
        union = nx.disjoint_union(union, copy)
    links = [(0, p) for p in union.nodes() if union.node[p]["peripheral"] == True]
    union.add_edges_from(links)
    return union
```

```python
for i in range(6):
    graph = replicate(graph)
```

```python
G = graph.copy()
result  = []
removed = []
coefficients = nx.clustering(G)

for key in sorted(coefficients, key=coefficients.get, reverse=True):
    G.remove_node(key)
    removed.append(key)
    try:
        giant = max(nx.connected_component_subgraphs(G), key=len)
        result.append({"fraction": len(removed)/N, "size": len(giant)})
    except:
        pass
    
df = pd.DataFrame(result)
df.plot(x="fraction", y="size", kind="line", legend=False)
plt.title("Removing the node with the \n highest clustering coefficient from the hierarchical model.")
plt.xlabel(r"Fraction nodes removed $(S/N)$", fontsize=16)
plt.ylabel(r"Size of largest component $(C)$", fontsize=16)
#plt.plot();
plt.savefig("picc.png")  
plt.close()
```
